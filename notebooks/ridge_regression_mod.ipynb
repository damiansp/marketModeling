```python
def median_fill(df):
    X = df.copy()
    
    for col in list(X):
        try:
            X[col][np.isnan(X[col])] = np.nanmedian(X[col])
        except:
            print('Could not fill column', col)
    return X
```

```python
X = median_fill(X)
X.head()
```
