```python
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
sym = ['PTON', 'PINS', 'COUR']
df = yf.download(sym, start='1990-01-01').drop('Volume', axis=1).rename(columns={'Adj Close': 'Value'})
if len(sym) == 1:
    df.columns = pd.MultiIndex.from_tuples([(x, sym[0]) for x in list(df)])
df.head()
```

```python
df['Date'] = df.index
df.index = range(df.shape[0])
df.head()
```

```python
df.loc[(2), ('Value', 'PTON')]
```

```python
n = df.shape[0]
for s in sym:
    df['LogValue', s] = np.log(df.Value[s])
    df['IntradayChange', s] = df.Close[s] / df.Open[s]
    df['DayToDayChange', s] = np.nan
    df['OvernightChange', s] = np.nan
    for day in range(1, n):
        df.loc[(day), ('DayToDayChange', s)] = (df.loc[(day), ('Value', s)] / df.loc[(day - 1), ('Value', s)])
        df.loc[(day), ('OvernightChange', s)] = (df.loc[(day), ('Open', s)] / df.loc[(day - 1), ('Close', s)])
df.tail()
```
